```python
def encode_categorical(df, categorical_columns):
    """
    Заменяет категориальные переменные на числовые значения.

    :param df: pandas DataFrame, содержащий данные.
    :param categorical_columns: список имен категориальных переменных.
    :return: pandas DataFrame с числовыми значениями вместо категориальных.
    """
    # Создаем копию DataFrame, чтобы не изменять оригина
    
    # Проходим по всем категориальным колонкам и преобразуем их с помощью LabelEncoder
    for column in categorical_columns:
        le = LabelEncoder()
        df[column] = le.fit_transform(df[column])
    
    return df


def get_categorical_columns(df):
    """
    Возвращает список категориальных переменных в DataFrame.

    :param df: pandas DataFrame, содержащий данные.
    :return: список имен категориальных переменных.
    """
    # Определяем категориальные переменные как те, у которых тип данных object или category
    categorical_columns = df.select_dtypes(include=['object', 'category']).columns.tolist()
    
    return categorical_columns
```

```python
categorical_columns = get_categorical_columns(df)
df_encoded = encode_categorical(df, categorical_columns)

df.info()
```

```python
def fill_nan_with_mode(df):
    """
    Заменяет все NaN значения в DataFrame на наиболее частое значение в каждом столбце.

    :param df: pandas DataFrame, содержащий данные.
    :return: pandas DataFrame с замененными NaN значениями.
    """
    # Проходим по каждому столбцу и заменяем NaN на наиболее частое значение
    for column in df.columns:
        if df[column].isna().any():
            mode_value = df[column].mode()[0]
            df[column].fillna(mode_value, inplace=True)
    
    return df


df = fill_nan_with_mode(df)
```

```python
df.to_csv('my_prepared.csv', sep=',', index=False, header=True, mode='w')
```
