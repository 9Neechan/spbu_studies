#### Dataset "babyboom"

```python
cols = ['Time of birth', 'Sex', 'Weight', 'Minutes after midnight']
df = pd.read_fwf('datasets/babyboom.dat.txt', header=None, widths=[8, 11, 8, 8], names=cols)

df.head()
```

```python
df.describe()
```

```python
print('Time of birth: ', statistics.variance(df['Time of birth']))
print('Sex: ', statistics.variance(df['Sex']))
print('Weight: ', statistics.variance(df['Weight']))
print('Minutes after midnight: ', statistics.variance(df['Minutes after midnight']))
```
