```python
import statistics
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set() 
```

### Задание 2
#### Dataset "babyboom"

```python
cols = ['Time of birth', 'Sex', 'Weight', 'Minutes after midnight']
df = pd.read_fwf('babyboom.dat.txt', header=None, widths=[8, 11, 8, 8], names=cols)

df.head()
```

```python
df.describe()
```

Выборочная дисперсия

```python
print('Time of birth: ', statistics.variance(df['Time of birth']))
print('Sex: ', statistics.variance(df['Sex']))
print('Weight: ', statistics.variance(df['Weight']))
print('Minutes after midnight: ', statistics.variance(df['Minutes after midnight']))
```
