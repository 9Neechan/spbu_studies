```python
import pandas as pd
import scipy.stats as st

df = pd.read_excel('datasets/surgery.xlsx')
df = df.dropna()

df.head()
```

```python
cols = [df['V_right_before'], df['V_left_before'], df['V_right_after'], df['V_left_after']]

def get_success(df):
    if df['V_right_before'] < df['V_right_after'] and df['V_left_before'] < df['V_left_after']:
        return 1
    else:
        return 0
    

df['succes'] = df.apply(lambda x: get_success(x), axis=1)
df['succes']
```
